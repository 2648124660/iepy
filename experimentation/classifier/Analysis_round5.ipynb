Data Loading
------------


Connect to server:

```python
from featureforge.experimentation.stats_manager import StatsManager
sm = StatsManager(1, "db_experiments", "goedel")
```

Load experiments for this round with diagnosed_reference.csv (md5 ab27daf09c202a2f788902bad0a7f6e4)

```python
config_version = u'5'
md5 = u'ab27daf09c202a2f788902bad0a7f6e4'
points = [p for p in sm.iter_results() if p[u'config_version'] == config_version and p[u'input_file_md5'] == md5]
len(points)
```

1. Average results across shufflings.
2. Group results by strategy (i.e. put together the different train_percentage).

```python
avgdata = average_stats(points)
data_bystrategy = group_by_strategy(avgdata)
```

Best Results Filtering
----------------------

First acceptability criteria for strategies:
- Average precision above 80% for all training sizes
- At least one training size has recall above zero

```python
accepted = {}
for key, xs in list(data_bystrategy.items()):
    if (all(x["results"]["precision_avg"] > 0.80 for x in xs) and  # Average precision above 80% for all training sizes
        any(x["results"]["recall_avg"] > 0 for x in xs)):          # At least one training size has recall above zero
        accepted[key] = xs

print("There are {} data points in total".format(len(points)))
print("After summarizing the averages of the 'cross-validation' runs there are {} data points".format(len(avgdata)))
print("The datapoints can be grouped into {} different 'strategies' (predictor configurations)".format(len(data_bystrategy)))
print("From all the strategies only {} meet the acceptability criteria".format(len(accepted)))
```

**Interesting...**

Print strategies and results (in no particular order):

```python
delfields = "dimensionality_reduction dimensionality_reduction_dimension feature_selection scaler sparse"
for pl in accepted.values():
    pprint_solution(pl[0], delfields)
    pprint_strategy_result(pl)
```

We find:
- kernel: always rbf
- gamma: always 0.0
- class weights and filtering frequency:
    - the six combinations: [false:10,true:1, false:1,true:1] x [2, 5, 10]
    - one extra combination (not good): (false:1,true:1, 1)

A couple of strategies have very bad F1.
Filter F1 < 50% and plot:

```python
label = lambda p: str(p[u'classifier_args'][u'class_weight'][u'false'])+'/'+str(p[u'feature_selection_dimension'])
solutions = [lp for lp in accepted.values() if max(p[u'results'][u'f1_avg'] for p in lp) > 0.5]
solutions = [(label(x[0]), x) for x in solutions]
show = dict(solutions)
myplot(show, "train_size", "f1_avg", dot="-o")
myplot(show, "train_size", "precision_avg", dot="-o")
```

Conclusions so far:
  - Use RBF with gamma = 0.0.
  - Class weights:
      - false:1,true:1 is better for the F1.
      - false:10,true:1 is better for precision.
      - Both are ok and better than any other.
  - Frequency filtering:
       - Anomaly: 2 is very bad with false:10,true:1.
       - For F1 there are good results for 2, 5 and 10 (5 and 10 slightly better).
       - For precision, 5 and 10 are the best among [1, 2, 5, 10].

Best 10 results according to average F1:

```python
len(data_bystrategy)
data_bystrategy2 = sorted(data_bystrategy.values(), key=lambda lp: max(p["results"]["f1_avg"] for p in lp), reverse=True)
delfields = "dimensionality_reduction dimensionality_reduction_dimension feature_selection scaler sparse"
for lp in data_bystrategy2[:10]:
    pprint_solution(lp[0], delfields)
    pprint_strategy_result(lp)
```

We find:
- kernel: always rbf
- gamma: always 0.0
- class weights and filtering frequency:
    - best: [false:1,true:10] x [1,2,5,10].
    - next: [false:1,true:1] x [2,5,10].
    - then another three: (false:10,true:1, 10), (false:0.1,true:1, 10), (false:10,true:1, 10)

Conclusions for this part:
  - Use RBF with gamma = 0.0.
  - Class weights:
      - Obvioulsy more weight to true gives more positives, usually worsening precision and improving recall.
      - false:1,true:10 improves the F1 (from 60% to 66%) but is very bad for precision (from 90% to 68%).


Conclusions
-----------

1. This set of features has the best results so far.
2. **RBF kernel** with **gamma 0.0** is giving the best results.
3. Best class weights:
    - **false:1,true:1**: good F1 and good precision.
    - **false:10,true:1**: slightly worse F1, slightly best precision.
    - false:1:true:10: slightly best F1, much worse precision.
4. Filtering frequency:
    - It looks like it is not as influential as the previous parameters.
    - No filtering at all (equivalently, filtering with frequency 1) seems to be the worst solution.
    - **5 and 10** seems to be the best frequency filters among [1, 2, 5, 10].
