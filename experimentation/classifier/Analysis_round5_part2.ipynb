Data Loading

Connect to server:

```python
from featureforge.experimentation.stats_manager import StatsManager
sm = StatsManager(1, "db_experiments", "goedel")
```

Load experiments for this round with causes_reference.csv (md5 bc75cee8fab98fdb295dd008b0a0fcf5)

```python
config_version = u'5'
md5 = u'bc75cee8fab98fdb295dd008b0a0fcf5'
points = [p for p in sm.iter_results() if p[u'config_version'] == config_version and p[u'input_file_md5'] != md5]
len(points)
```

1. Average results across shufflings.
2. Group results by strategy (i.e. put together the different train_percentage).

```python
avgdata = average_stats(points)
data_bystrategy = group_by_strategy(avgdata)
```

Best Results Filtering
----------------------

First acceptability criteria for strategies:
- Average precision above 80% for all training sizes
- At least one training size has recall above zero

```python
accepted = {}
for key, xs in list(data_bystrategy.items()):
    if (all(x["results"]["precision_avg"] > 0.80 for x in xs) and  # Average precision above 80% for all training sizes
        any(x["results"]["recall_avg"] > 0 for x in xs)):          # At least one training size has recall above zero
        accepted[key] = xs

print("There are {} data points in total".format(len(points)))
print("After summarizing the averages of the 'cross-validation' runs there are {} data points".format(len(avgdata)))
print("The datapoints can be grouped into {} different 'strategies' (predictor configurations)".format(len(data_bystrategy)))
print("From all the strategies only {} meet the acceptability criteria".format(len(accepted)))
```

**Interesting...**

Print strategies and results (in no particular order):

```python
delfields = "dimensionality_reduction dimensionality_reduction_dimension feature_selection scaler sparse"
for pl in accepted.values():
    pprint_solution(pl[0], delfields)
    pprint_strategy_result(pl)
```

We find:
- kernel: both rbf and poly
- gamma: always 0.0
- class weights: [false:10,true:1, false:1,true:0.1, false:1,true:1]
- filtering frequencies: 1 and 2

Very bad recall and F1 in general.
Plots:

```python
label = lambda p: str(p[u'classifier_args'][u'kernel'])+'/'+str(p[u'classifier_args'][u'class_weight'])+'/'+str(p[u'feature_selection_dimension'])
#solutions = [lp for lp in accepted.values() if max(p[u'results'][u'f1_avg'] for p in lp) > 0.5]
solutions = [(label(lp[0]), lp) for lp in accepted.values()]
show = dict(solutions)
myplot(show, "train_size", "f1_avg", dot="-o")
myplot(show, "train_size", "precision_avg", dot="-o")
```

Conclusions so far:
  - Again, gamma = 0.0.
  - Results are bad but comparable to those on round 3.
  - Precision is very high and recall is very low because there are almost no guesses.
  - The best choice seems to be RBF with gamma=0.0, class weight=false:10,true:1, and frequency filtering=2.
  - Unlike round 5 part 1, we don't find frequency filterings 5 and 10.

Best results according to average F1 (only F1 > 50%):

```python
len(data_bystrategy)
data_bystrategy2 = sorted(data_bystrategy.values(), key=lambda lp: max(p["results"]["f1_avg"] for p in lp), reverse=True)
data_bystrategy3 = [lp for lp in data_bystrategy2 if max(p["results"]["f1_avg"] for p in lp) > 0.5]
delfields = "dimensionality_reduction dimensionality_reduction_dimension feature_selection scaler sparse"
for lp in data_bystrategy3:
    pprint_solution(lp[0], delfields)
    pprint_strategy_result(lp)
```

We find:
- kernel: always rbf
- gamma: always 0.0
- class weight: always false:1,true:10
- filtering frequency: 5, 10 and 2.

Conclusions for this part:
  - Use RBF, gamma = 0.0.
  - Class weight false:1,true:10 gives the best F1 but with low precision.

Comparison with Part 1
----------------------

Find results for best 4 strategies found in Part 1.

```python
label = lambda p: str(p[u'classifier_args'][u'kernel'])+'/'+str(p[u'classifier_args'][u'class_weight'])+'/'+str()

accepted = {}
for key, xs in list(data_bystrategy.items()):
    p = xs[0]
    kernel = p[u'classifier_args'][u'kernel']
    gamma = p[u'classifier_args'][u'gamma']
    class_weight = p[u'classifier_args'][u'class_weight']
    frequency_filtering = p[u'feature_selection_dimension']
    if kernel == u'rbf' and gamma == 0.0 and class_weight in [{u'false': 10, u'true': 1}, {u'false': 1, u'true': 1}] and frequency_filtering in [5, 10]:
        accepted[key] = xs

delfields = "dimensionality_reduction dimensionality_reduction_dimension feature_selection scaler sparse"
for pl in accepted.values():
    pprint_solution(pl[0], delfields)
    pprint_strategy_result(pl)
```

- All of them have acceptable precision. 
- As always, class weight gives:
    - false:1,true:1 better F1, slightly worse precision.
    - false:10,true:1: slightly worse F1, better precision.
- No conclusions for filtering frequency.

Conclusions
-----------

With the "causes" relation, this set of features works more or less like the features used in round 3.
The results are not good in general, but the 'causes' dataset is too small to be trusted.
