```python
##
## The code below defines helper functions, retrieves experiment data from the server
## and prints a list of available data chunks from the server.
##

#%pylab --no-import-all inline

from collections import defaultdict
import pprint
import copy
import json
import time

from featureforge.experimentation.stats_manager import StatsManager
from featureforge.experimentation.utils import DictNormalizer


def pprint_solution(d):
    delfields = "results classifier_config database_hash data_shuffle_seed config_version input_file_md5 marshalled_key git_info".split()
    delfields.extend("experiment experiment_status".split())
    d = copy.deepcopy(d)
    for name in delfields:
        del d[name]
    pprint.pprint(d)

##
## Load data from experiments server
##

ROUND = 6

sm = StatsManager(1, "db_loop_experiments_results", "goedel")
alldatapoints = list(sm.data.find({sm.experiment_status: sm.STATUS_SOLVED, 'config_version': str(ROUND)}))

def rday(r):
    t = r[u'results'][u'start_time']
    t = time.gmtime(t)
    return (t.tm_year, t.tm_mon, t.tm_mday)

def rdatasize(r):
    return r[u'results'][u'dataset_size']

def rkey(r):
    return (r[u'config_version'],
            r[u'git_info'],
            r[u"input_file_md5"],
            tuple(r[u"database_hash"])
            )

groupeddata = defaultdict(list)
for r in alldatapoints:
    del r["booked_at"]
    del r["_id"]
    groupeddata[rkey(r)].append(r)

alldata = sorted(groupeddata.values(), key=lambda x: rday(x[0]))
print("\n id date\truns\tdataset size")
for i, xs in enumerate(alldata):
    date = "/".join("{:02}".format(x) for x in rday(xs[0]))
    print("{: 3} {}\t{}\t{}".format(i, date, len(xs), rdatasize(xs[0])))
```

```python
from IPython.display import display, HTML
import pygal

def get_key(point, key, default=None):
    key_path = key.split('.')
    value = point
    for k in key_path:
        if not isinstance(value, dict):
            # we'll use k as index
            value = value[int(k)]
        else:
            value = value.get(k, default)
    return value


def avg_var(values):
    avg = sum(values) / float(len(values))
    if len(values) == 1:
        var = 0
    else:
        var = sum((y - avg) ** 2 for y in values) / (len(values) - 1)
    return avg, var


def galplot(xs, x_path, y_path, exclude_criteria=None, labeler=None, group_by=None):  
    if callable(exclude_criteria):
        xs = exclude_criteria(xs)
        
    if callable(group_by):
        xxs = group_by(xs)
    else:
        xxs = dict([(nr, [x]) for nr, x in enumerate(xs, 1)])
    
    if callable(x_path):
        x_title = getattr(x_path, 'func_name', 'x')
        get_x = x_path
    else:
        x_title = x_path.split('.')[-1]
        get_x = lambda pt: get_key(pt, x_path)
    
    if callable(y_path):
        y_title = getattr(y_path, 'func_name', 'y')
        get_y = y_path
    else:
        y_title = y_path.split('.')[-1]
        get_y = lambda pt: get_key(pt, y_path)
    
    xy_chart = pygal.XY(stroke=True,
                        x_title=x_title,
                        y_title=y_title,
                        disable_xml_declaration=True)
    
    for group, points in xxs.items():
        if labeler:
            lbl = str(labeler(points[0]))
        else:
            lbl = '%s' % group
        xy_chart.add(lbl, [(get_x(pt), get_y(pt)) for pt in points])

    display(HTML("""
    <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
    <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
    """))
    display(HTML(xy_chart.render()))
```

```python
## Choose the data points to use from all available in the mongo server in the line below
data = alldata[1]

## Expand each data point according to each round
xp_data = []
for exp in data:
    exp = copy.deepcopy(exp)
    r = exp['results']
    answers = r.pop('answers_given')
    answers_per_round = exp[u'answers_per_round']
    rounds_stuff = r.pop('learning_numbers')
    for round_nr, learning_stats in enumerate(rounds_stuff, 1):
        exp = copy.deepcopy(exp)
        r = exp['results']
        r['answers_given'] = answers[:round_nr*answers_per_round]
        r['learning_stats'] = learning_stats
        r['round'] = round_nr
        xp_data.append(exp)

print len(xp_data)
```

```python
print("There are {} data points in total".format(len(data)))

final_kn = 'results.learning_numbers.-1.knownledge'

def keep_only_those_above(xs):
    rst =  [pt for pt in xs if
            #len(get_key(pt, 'results.answers_given')) < 100 and
            get_key(pt, final_kn + '.recall') >= 0.3 and
            get_key(pt, final_kn + '.precision') >= 0.5
            ]
    print 'left %s of %s points' % (len(rst), len(xs))
    return rst

def sort_by(xs, key_path, reverse=False):
    return sorted(xs, key=lambda pt: get_key(pt, key_path), reverse=reverse)

def labeler(pt):
    hs = get_key(pt, 'results.learning_numbers.-1.human_learnt_size')
    ks = get_key(pt, 'results.learning_numbers.-1.knownledge_size')
    ans = get_key(pt, 'results.answers_given')
    return (hs, ks, ans.count(1.0) / float(len(ans)))


galplot(data,
       final_kn + '.recall',
       final_kn + '.precision', 
       exclude_criteria=lambda xs: sort_by(keep_only_those_above(xs), final_kn + '.f1', True),
       labeler=labeler
      )
```

```python
ml_path = 'results.learning_stats.machine_learnt'
ml_pre = 'results.learning_stats.machine_accuracy'
ml_rec = 'results.learning_stats.machine_recall'  # is generated on next ipython cell
ml_f1 = 'results.learning_stats.machine_f1'       # is generated on next ipython cell
kn_pre = 'results.learning_stats.knownledge.precision'
kn_rec = 'results.learning_stats.knownledge.recall'
kn_f1 = 'results.learning_stats.knownledge.f1'
def keep_ones_with_machine_learnt(xs):
    result = [pt for pt in xs if get_key(pt, ml_path) > 0]
    return result

def results_cmp(el):
    # returns the pieces of results to use for a comparison
    return map(lambda x: get_key(el, x), [ml_path, ml_pre, kn_pre, kn_rec, 'results.answers_given'])

def group_by_hash(xs):
    xxs = defaultdict(list)
    for x in xs:
        xxs[x[u'marshalled_key']].append(x)
    print 'Created %s groups from %s initial elements' % (len(xxs), len(xs))
    # Will now remove groups with very bad general numbers
    for group, elems in list(xxs.items()):
        # If final prec and recall are not both higher than P & R, go home
        P = 0.55
        R = 0.15
        if get_key(elems[-1], kn_pre) < P or get_key(elems[-1], kn_rec) < R:
            del xxs[group]
            continue           
        # If machine accuracy is never higher than MA, go home
        MA = 0.2
        if all(get_key(el, ml_pre) < MA for el in elems):
            del xxs[group]
            continue
        # If machine learnt is never higher than ML, go home
        ML = 10
        if all(get_key(el, ml_path) < ML for el in elems):
            del xxs[group]  
            continue
        # If N elems have proposed more than MML, go home
        N = 10
        MML = 1200
        if len([el for el in elems if get_key(el, ml_path) > MML]) > N:
            del xxs[group]  
            continue
        # On each round, (after X rounds, for stability) F1 shall not degrade more than F1_degr, otherwise go home
        F1_degr = 0.05  # the higher, the more tolerant
        P_degr = 0.05   # the higher, the more tolerant
        is_fine = True
        N = 4
        prev = elems[N]
        for el in elems[N+1:]:
            if get_key(el, kn_f1) + F1_degr < get_key(prev, kn_f1):
                is_fine = False
                break
            elif get_key(el, kn_pre) + P_degr < get_key(prev, kn_pre):
                is_fine = False
                break
            else:
                prev = el
        if not is_fine:
            del xxs[group]
            continue
        
        # If IEPY F1 did not grow up F1Min, go home
        F1Min = 0.15
        if get_key(elems[-1], kn_f1) < F1Min + get_key(elems[0], kn_f1):
            del xxs[group]  
            continue           

    print 'Left %s groups after filters' % len(xxs)
    new_xxs = {}
    for group, elems in xxs.items():
        elems = sorted(elems, key=lambda e: e['results']['round'])
        for n_g, n_elems in list(new_xxs.items()):
            if [results_cmp(e) for e in elems] == [results_cmp(e) for e in n_elems]:
                #print 'Hiding %s behind %s'  % (group, n_g)
                break  # already considered
        else:
            new_xxs[group] = elems
        
    print 'And after regrouping, left %s' % len(new_xxs)
    return new_xxs

def len_answers(pt):
    return len(get_key(pt, 'results.answers_given'))

def f1_star(pt):
    return get_key(pt, ml_f1) * get_key(pt, kn_f1)


grouped_data = group_by_hash(xp_data)
```

```python
total_positives = 208  # hardcoded number of positive evidences on the reference corpus
def compute_machine_f1(item):
    human_positives = get_key(item, 'results.learning_stats.human_learnt_size')
    missing_evidences = float(total_positives - human_positives)
    machine_guesses = get_key(item, ml_path)
    machine_accuracy = get_key(item, ml_pre)
    machine_ok = machine_guesses * machine_accuracy
    machine_recall = machine_ok / missing_evidences
    try:
        machine_f1 = 2 * (machine_accuracy * machine_recall) / (machine_accuracy + machine_recall)
    except ZeroDivisionError:
        machine_f1 = 0.0
    item['results']['learning_stats']['machine_recall'] = machine_recall
    item['results']['learning_stats']['machine_f1'] = machine_f1

for dds in grouped_data.values():
    for d in dds:
        compute_machine_f1(d)
```

```python
def split_by_policies(data, what):
    policies = defaultdict(dict)
    for k, d in data.items():
        el = d[0]
        policies[get_key(el, what)][k] = d
    print 'There are %s policies, for a total of %s points' % (
            len(policies),
            sum(len(v) for v in policies.values())
            )
    return policies
```
